```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import urllib
import requests
import matplotlib.pyplot as plt
import os
from scrape import get_sermons_timed
speaker = 'adam_clarke'
print(os.getcwd())
#os.chdir('./machine_churching')
#os.mkdir(speaker)
get_sermons_timed(speaker, 'http://www.sermonindex.net/modules/articles/index.php?view=category&cid=438&page=')
```

```python
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/O/i/x/Y/q/P/yellow-house-hi.png', stream=True).raw))
```

```python
def generate_wordcloud(words, mask):
    word_cloud = WordCloud(width = 1024, height = 1024, background_color='white', stopwords=STOPWORDS, mask=mask).generate(words)
    word_cloud.to_file('./wordclouds/' + speaker + '_house.png')
    plt.figure(figsize=(10,8),facecolor = 'white', edgecolor='blue')
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.tight_layout(pad=0)
    plt.show()
```

```python
contents = ''
os.chdir('./'+ speaker)
for sermon in os.listdir():
    f=open(sermon, "r")
    if f.mode == 'r':
        contents +=f.read()
os.chdir('..')
```

```python
generate_wordcloud(contents, mask)
```
