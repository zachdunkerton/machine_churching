```python
%matplotlib inline
from sklearn.cluster import KMeans
import pickle
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
from sklearn.manifold import TSNE
with open('./embeddings/doc2vec_embeddings.pkl', 'rb') as handle:
    embeddings = pickle.load(handle)
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, max_iter=1000).fit(embeddings)
kmeans_data = kmeans.transform(embeddings)
```

```python
pca = PCA(n_components = 50)
pca_reduction = pca.fit_transform(embeddings)
```

```python
graph_data = TSNE(n_components=2).fit_transform(pca_reduction)
```

```python
fig = plt.figure(figsize=(20, 13), dpi= 80, facecolor='w', edgecolor='k')
ax1 = fig.add_subplot(111)
ax1.scatter(graph_data[:][:,0],graph_data[:][:,1], c = kmeans.labels_)
plt.legend(prop={'size': 15})
plt.show()
```
